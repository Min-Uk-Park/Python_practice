```python
import os,sys,re
import csv
import pandas as pd
from folium.plugins import MarkerCluster
import folium
from haversine import haversine # 위도, 경도 이용하여 구 상에서 두 점 사이의 직선거리 구하기(default 단위는 km)
import requests

os.chdir(f"c:/Users/LG/Desktop/") # C:\Users\LG

# print(os.listdir()) # 전라북도_전주시_주차장정보_20230704.csv



for file in os.listdir() :
    if (re.match(f"전라북도_전주시_주차장정보_20230704.csv",file)) :
        df = pd.read_csv(file,thousands=",",encoding="cp949")
    
    
    
```

```python
m = folium.Map([35.8242238,127.1479532],zoom_start=17,width=1200,height=500)


mc = MarkerCluster().add_to(m)

folium.Marker(
    location=[35.8242238,127.1479532],
    tooltip="전주 시청",
    popup= "jeonju",
    ).add_to(mc)

for lat,long,name,adrr in  zip(df['위도'],df['경도'],df['주차장명'],df['소재지도로명주소']):
    folium.Marker(
    location=[lat,long],
    tooltip= name,
    
    icon= folium.Icon(icon='car',prefix = 'fa',color = 'green'),
    popup= folium.Popup(html = f'<br>{adrr}</br>',max_width=300)
    
    ).add_to(mc)
m
```
