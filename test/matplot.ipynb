```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # matplotlib.pyplot은 matplotlib패키지에 있는 하위 패키지

plt.rcParams['font.family']='Malgun Gothic' # 차트에서 한글 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False  # 차트에서 -기호 깨짐 방지
```

```python
plt.plot([1, 4, 9, 16]) # x tick은 자동으로 [0,1,2,3]이 된다.
plt.title("차트 연습")
plt.show()
```

```python
plt.title('차트 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("x축 제목")
plt.ylabel("y축 제목")
plt.show()
```

```python
font1 = { 'size': 24, 'color' :  'black', 'weight' : 'bold'}
font2 = { 'size': 13, 'weight': 'bold', 'color':  'darkred'}
font3 = { 'size': 13,'family': 'Malgun Gothic', 'weight': 'light', 'color':  'darkblue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('차트 제목', fontdict=font1) # 글자모양은 fontdict 인자를 사용하여 지정한다.
plt.xlabel('x축 제목', fontdict=font2)
plt.ylabel('y축 제목', fontdict=font3) # 한글 글꼴은 Malgun Gothic 만 지원한다. 
plt.show()
```

```python
X = np.linspace(-5, 5, 300, retstep="True") # 간격이 얼마인지도 반환
X = np.linspace(-5, 5, 300)
X
C = np.cos(X)    # np.cos(np.pi) 는 -1 입니다. np.cos(0) 은 1 입니다. 
C
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
# plt.xticks(np.linspace(-5, 5, 5))
plt.xticks([-np.pi,0,np.pi])
plt.yticks([-1, 0, +1])
plt.grid(True) # 그리드 가 보입니다. plt.grid(False) 는 그리드 감추기 입니다.
plt.show()
```

```python
np.random.seed(0)
f1 = plt.figure(figsize=(10, 2))
plt.plot(np.random.randn(100))
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1) # ax1으로 지정해주는 이유는 ax1의 그래프 값을 바꿀 때 ax1. 형태로 지정가능하기 때문이다!
plt.plot(x1, y1, 'yo-',label = 'cos')
# plt.title('A tale of 2 subplots')
ax1.set_title('axs1입니다.')
plt.ylabel('Damped oscillation')  # 감쇠진동
plt.text(3,0.5,'안녕하세요') # x(3), y(0.5) 위치에 지저한 텍스트 입력.
plt.legend(ncol = 1) # 범례지정

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-',label = 'cos2')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.legend(ncol = 3)

plt.tight_layout() # 그래프 겹치는 것 방지!!!
plt.show()
```
