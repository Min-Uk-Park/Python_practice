```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # matplotlib.pyplot은 matplotlib패키지에 있는 하위 패키지

plt.rcParams['font.family']='Malgun Gothic' # 차트에서 한글 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False  # 차트에서 -기호 깨짐 방지
```

```python
plt.plot([1, 4, 9, 16]) # x tick은 자동으로 [0,1,2,3]이 된다.
plt.title("차트 연습")
plt.show()
```

```python
plt.title('차트 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("x축 제목")
plt.ylabel("y축 제목")
plt.show()
```

```python
font1 = { 'size': 24, 'color' :  'black', 'weight' : 'bold'}
font2 = { 'size': 13, 'weight': 'bold', 'color':  'darkred'}
font3 = { 'size': 13,'family': 'Malgun Gothic', 'weight': 'light', 'color':  'darkblue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('차트 제목', fontdict=font1) # 글자모양은 fontdict 인자를 사용하여 지정한다.
plt.xlabel('x축 제목', fontdict=font2)
plt.ylabel('y축 제목', fontdict=font3) # 한글 글꼴은 Malgun Gothic 만 지원한다. 
plt.show()
```

```python
plt.title("'rs--' 스타일 차트 ")
# plt.plot([10, 20, 30, 40], [1, 4, 9, 16],'rs--')
# plt.plot([10, 20, 30, 40], [1, 8, 27, 48], 'bD-')
# 아래처럼 묶을 수 있다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs-',[10,20,30,40], [1, 8, 27, 48], 'bD-')
# [10, 20, 30, 40], [1, 4, 9, 16], 'rs--' 3가지 인자가 하나의 그래프를 그려줌
# [10, 20, 30, 40], [1, 8, 27, 48], 'bD--' 3가지 인자가 또 하나의 그래프를 그려줌
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50) # plt.axis([0, 50, -10, 30]) x축범위 0~50, y축범위 -10~30
plt.ylim(-10, 30)

plt.show()
```

```python
X = np.linspace(-5, 5, 300, retstep="True") # 간격이 얼마인지도 반환
X = np.linspace(-5, 5, 300)
X
C = np.cos(X)    # np.cos(np.pi) 는 -1 입니다. np.cos(0) 은 1 입니다. 
C
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
# plt.xticks(np.linspace(-5, 5, 5))
plt.xticks([-np.pi,0,np.pi])
plt.yticks([-1, 0, +1])
plt.grid(True) # 그리드 가 보입니다. plt.grid(False) 는 그리드 감추기 입니다.
plt.show()
```

```python
# x ticks, y ticks 이해하기

plt.title('ticks 이해')
# plt.plot([1,2,3,4,5,6]) # 기본 x축 값들은 0,1,2,3,4~ 로 설정된다.

plt.plot([10,20,30,40,50],[100,400,900,1600,2500])
plt.xlim(0,60) # x축 범위 설정
plt.xticks([10,30,50]) # x축이 숫자로 되어있기 때문에 xticks에 해당하는 숫자값에 대응하는 숫자만 보이게 된다.
# 그러나
# plt.xticks(['2020년','2021년','2023년']) # 10, 30, 50 부분에 각각 '2020년','2021년','2023년'을 넣을 때는 error가 난다!!

plt.xticks([10,30,50],['2020년','2021년','2023년']) # 어느 숫자 값에 문자(또는 문자열)을 넣을지 대응되게 한다!
```

```python
# 막대 그래프 두 개 동시에 그려보기

plt.bar([10,20,30,40,50],[10,20,30,40,50],width=2.5,color='r')
# plt.bar([10,20,30,40,50],[10,20,30,40,50],width=3)

# x축 값들을 리스트로 표현해야 하기 때문에 편하게 numpy의 arange함수를 사용해준다.
# plt.bar(np.arange(10,51,10,),[10,20,30,40,50],width=2.5,color ='b',)

# 그래프가 겹쳐 보이기 때문에 x축 값들을 막대 너비보다 큰 값을 더해 이동시킨다.
plt.bar(np.arange(10,51,10,)+4,[10,20,30,40,50],width=2.5,color ='b',)

# 이때 x축 값이 왼쪽으로 치우쳐져 있기 때문에 중앙으로 옮겨준다. 
plt.xticks(np.arange(10,51,10,)+2,[10,20,30,40,50])
```
